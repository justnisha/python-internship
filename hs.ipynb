```python
import pandas as pd

# Load the dataset (replace 'your_dataset.csv' with your actual file path)
df = pd.read_csv('housing.csv')

# Check for missing values before imputation
print("Missing values before imputation:\n", df.isnull().sum())

# Fill missing values in numerical columns with the mean
df.fillna(df.mean(numeric_only=True), inplace=True)  # Use median: df.fillna(df.median(numeric_only=True), inplace=True)

# Check for missing values after imputation
print("\nMissing values after imputation:\n", df.isnull().sum())
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('housing.csv')

# Fill missing numerical values with the mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# Fill missing categorical values with mode or 'Unknown'
categorical_cols = df.select_dtypes(include=['object']).columns
df.fillna({col: df[col].mode()[0] if not df[col].mode().empty else "Unknown" for col in categorical_cols}, inplace=True)

# Check for missing values after imputation
print("\nMissing values after imputation:\n", df.isnull().sum())
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('housing.csv')

# List of columns to drop (replace with actual column names)
columns_to_drop = ['Condition2']  # Example: ['ID', 'Unnamed: 0']

# Drop columns
df.drop(columns=columns_to_drop, inplace=True)

# Drop columns with too many missing values (optional)
threshold = 0.5  # If more than 50% values are missing, drop the column
df.dropna(axis=1, thresh=int(threshold * len(df)), inplace=True)

# Check remaining columns
print("\nRemaining Columns:\n", df.columns)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
df = pd.read_csv('housing.csv')

# Select numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Apply MinMaxScaler and update the dataframe
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Check the normalized data
print("\nNormalized Data:\n", df.head())
```

```python
# Save the cleaned dataset to a new CSV file
df.to_csv('cleaned_dataset.csv', index=False)

print("Cleaned dataset saved as 'cleaned_dataset.csv' successfully!")
```
